```python
import exifread

with open("./portfolio_data/a_slice_of_urban_life/goingToSproutsAtGoldenTime-20251009-183334.JPG", "rb") as f:
    tags = exifread.process_file(f, details=False)

print("Aperture:", tags.get("EXIF FNumber"))
print("ISO:", tags.get("EXIF ISOSpeedRatings"))
print("Focal Length:", tags.get("EXIF FocalLength"))
print("Shutter Speed:", tags.get("EXIF ExposureTime"))
```
